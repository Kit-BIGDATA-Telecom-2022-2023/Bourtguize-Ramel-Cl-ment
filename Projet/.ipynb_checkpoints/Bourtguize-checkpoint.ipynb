## Scrapping et préparation donnés relatives aux charactéristiques technique des bateau


```python
url='https://www.vendeeglobe.org/fr/glossaire'
r=rq.get(url)
html=r.content.decode('utf-8')
soup=BeautifulSoup(html)
l=soup.find_all('ul',{'class' : "boats-list__popup-specs-list"})
carac=[]
for m in l:
    p=re.compile(r'(.+)\n')
    ll=p.findall(m.text)
    spec={}
    for line in ll :
        p=re.compile('( : )')
        k=p.search(line)
        spec[line[0:k.span()[0]]]=line[k.span()[1]:len(line)]
    carac.append(spec)
Cara=pd.DataFrame(carac)
```

```python
Cara[Cara['Numéro de voile'].isna()]
```

```python
Cara.iloc[26]['Numéro de voile']='FRA 59'
#la ligne 16, le skipper est francois-guiffant mais il n'apparait pas dans les classement excel, c'est donc la ligne en trop 
#(33 ligne dans les excel contre 34 dans le data frame)
Cara1=Cara.drop(labels=16,axis=0)
Cara1=Cara1.reset_index(drop=True)
```

```python
p=re.compile('[0-9]+')
Cara1['Numéro de voile'].apply(lambda x : p.search(x).group()).nunique()
```

On peut donc faire une jointure sur le numéro de voile indépendament du pays, afin de préparer cela, on crée un colonne avec le numéro sans le tag du pays

```python
Cara1['Num voile']=Cara1['Numéro de voile'].apply(lambda x : p.search(x).group())
```

## rapprochement des deux tables
dans un soucis d'optimisation on ne ferat pas la jointure entre les tables des classement et celle des caractérisitque car cela donnerais une datafranme trop voluminieux avec beaucoup de redondance. On effectuera des jointure en fonction des besoin de traitements. On se propose tout de même d'effectuer une fois cette jointure pour une table de classement, uniquement pour démontrer que cette opération est maitrisé.

```python
#On doit passer un multiindex sur les colonnes pour pouvoir faire un join avec la table des classement.
#On peut donc se demander restrospectivement si le choix du MultiIndex était judicieux.
CaractJoin=Caract.copy()
CarCol=Caract.columns.values
CarJoinCol=[(l,l) for l in CarCol]
CaractJoin.columns=pd.MultiIndex.from_tuples(CarJoinCol)
CaractJoin
```
